## Importing necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

## Reading the dataset

```python
data = pd.read_csv('dataset.csv')
```

## 1/Generate Bar Chart for Gender Distribution

### Count the number of occurrences of each gender

```python
gender_counts = data['gender'].value_counts()
```

### Create a bar chart

```python
plt.figure(figsize=(8, 6))
plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('count')
plt.show()
```

## 2/Create a histogram for Age

##### the plt.hist() function does this automatically by counting how many values of age fall within each bin and then plotting that count (frequency) as the height of the bar.

```python
plt.figure(figsize=(8, 5))
plt.hist(data['age'], bins=8, color='green', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
